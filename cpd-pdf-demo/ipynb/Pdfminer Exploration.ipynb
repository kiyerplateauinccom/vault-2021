```python

import os

# cpd-pdf-demo/data/pdf/eanaobck_11_3.pdf
home_dir = dict(os.environ)['HOME']
pdf_dir = f'{home_dir}/cpd-pdf-demo/data/pdf'
pdf_paths_list = []
for file_name in os.listdir(pdf_dir):
    if file_name.endswith('.pdf'):
        file_path = os.path.join(pdf_dir, file_name)
        pdf_paths_list.append(file_path)
```

```python

import random
import sys
from pdfminer.pdfparser import PDFParser
from pdfminer.pdfdocument import PDFDocument
from pdfminer.pdftypes import resolve1

def get_pdf_document(file_path):
    fp = open(file_path, 'rb')
    parser = PDFParser(fp)
    doc = PDFDocument(parser)
    
    return doc
def print_fields(file_path):
    doc = get_pdf_document(file_path)
    acroform_dict = resolve1(doc.catalog['AcroForm'])
    fields_list = acroform_dict['Fields']
    for i in fields_list:
        field = resolve1(i)
        name = field.get('T')
        if name is not None:
            name = name.decode()
        value = field.get('V')
        if value is not None:
            value = value.decode()
        print('{0}: {1}'.format(name, value))
```

```python

file_path = random.choice(pdf_paths_list)
doc = get_pdf_document(file_path)
while 'AcroForm' not in doc.catalog:
    file_path = random.choice(pdf_paths_list)
    doc = get_pdf_document(file_path)
print_fields(file_path)
```
